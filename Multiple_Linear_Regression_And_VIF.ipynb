```python
surg_data=df2["Surg - Med"]==0
df2=df2.loc[surg_data]
df2
```

```python
import statsmodels.api as sm
x = df2.loc[:, ['Age', 'Severity', 'Anxiety']].copy()
y = df2['Satisfaction'].copy()
print(x)
print(y)
```

```python
x = sm.add_constant(x)
model1 = sm.OLS(y, x)
results1 = model1.fit()
print(results1.summary())
print("\n Fitted Values:\n")
y_pred = results1.fittedvalues.round(2)
print(y_pred)
```
